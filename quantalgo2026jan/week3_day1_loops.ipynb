```python
import random

price = 100
target_price = 105
max_days = 30  # Safety limit (we won't wait forever)
day = 0

while price < target_price and day < max_days:
    day += 1
    # Random daily change between -$2 and +$3
    price_change = random.uniform(-2, 3)
    price += price_change
    print(f"Day {day}: Price = ${price:.2f} (change: {price_change:+.2f})")

if price >= target_price:
    print(f"\nâœ… Target of ${target_price} reached on day {day}!")
else:
    print(f"\nâ° Max days reached. Final price: ${price:.2f}")
```

```python
import random

# Simulate waiting for a stop-loss to trigger
entry_price = 150
stop_loss = 140  # Exit if price drops to $140
day = 0
max_days = 20

print(f"Entered at ${entry_price}, Stop-loss at ${stop_loss}\n")

current_price = entry_price

while current_price > stop_loss and day < max_days:
    day += 1
    daily_change = random.uniform(-5, 3)  # Market can drop faster than it rises
    current_price += daily_change
    
    print(f"Day {day}: ${current_price:.2f} (change: {daily_change:+.2f})")

print("\n" + "="*40)
if current_price <= stop_loss:
    loss = ((current_price - entry_price) / entry_price) * 100
    print(f"âŒ Stop-loss hit on day {day}")
    print(f"Loss: {loss:.2f}%")
else:
    gain = ((current_price - entry_price) / entry_price) * 100
    print(f"âœ… Position held for {day} days")
    print(f"Current P&L: {gain:+.2f}%")
```

```python
account_values = [10000, 10500, 10200, 9800, 9500, 10100, 10800]
peak = account_values[0]  # Start with first value as peak
max_drawdown = 0          # No drawdown yet

for value in account_values:
    # Update peak if current value is higher
    if value > peak:
        peak = value
    
    # Calculate drawdown from current peak
    drawdown = (peak - value) / peak
    
    # Track the maximum drawdown seen
    if drawdown > max_drawdown:
        max_drawdown = drawdown
    
    print(f"Value: ${value:,}, Peak: ${peak:,}, Drawdown: {drawdown:.2%}")

print(f"\nðŸ“‰ Maximum Drawdown: {max_drawdown:.2%}")
```

```python
# Your trading account over 2 months
my_account = [5000, 5200, 5100, 4800, 4500, 4700, 5000, 5300, 5100, 5500]

peak_value = my_account[0]
worst_drawdown = 0

print("Day-by-Day Analysis:")
print("="*60)

for day, balance in enumerate(my_account, start=1):
    # Update peak if we hit a new high
    if balance > peak_value:
        peak_value = balance
    
    # Calculate current drawdown
    current_dd = (peak_value - balance) / peak_value
    
    # Track the worst drawdown
    if current_dd > worst_drawdown:
        worst_drawdown = current_dd
        worst_day = day
    
    print(f"Day {day:2}: ${balance:,} | Peak: ${peak_value:,} | DD: {current_dd:6.2%}")

print("="*60)
print(f"ðŸ”´ Worst Drawdown: {worst_drawdown:.2%} (occurred on Day {worst_day})")
print(f"ðŸ’° Final Balance: ${my_account[-1]:,}")
print(f"ðŸ“ˆ Overall Return: {((my_account[-1] - my_account[0]) / my_account[0]):.2%}")
```
